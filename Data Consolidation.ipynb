```python
import pandas as pd
import glob
import os
```

```python
dataset = ['training', 'test', 'validation']
work = ['KMeans', 'PageRank', 'SGD', 'tensorflow', 'web_server']
```

```python
path = r'data/'
count = {'KMeans':0, 'PageRank':0, 'SGD':0}
for w in work[-2:]:
    if w not in count:
        count[w] = 0
    for ds in dataset[-2:]:

        ind = 0
        
        path_new = path + w + '/' + ds
        all_files = glob.glob(os.path.join(path_new, "*.csv"))
        total = None
        for filename in all_files:
            ind+=1
            df = pd.read_csv(filename, header = 0, index_col = None)
            del df['index']
            df['job'] = ind
            if total is not None:
                total = pd.concat([total,df])
            else:
                total = df
        if ds=='test':
            count[w]+=ind
            
        total.to_csv(w + '_' + ds + '_jb' + '.csv', index = False)
print(count)
```

```python
path = r'/data'
total_lines = 0
for ds in dataset:
    all_files = glob.glob(os.path.join(path, "*_" + ds + ".csv"))
    total = []
    for filename in all_files:
        df = pd.read_csv(filename, header = 0, index_col = None)
        total_lines += df.shape[0]
        work_name = filename.split('_')[0]
        ind = work.index(work_name)
        df['work']=ind+1
        total.append(df)
    blah = pd.concat(total)
    blah.to_csv(ds + '.csv', index = False)
assert total_lines == blah.shape[0]
```
