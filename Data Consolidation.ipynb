```python
import pandas as pd
import glob
import os
```

```python
dataset = ['training', 'test', 'validation']
work = ['KMeans', 'PageRank', 'SGD']
path = r''
```

```python
total_lines = 0
for ds in dataset:
    total_lines = 0
    all_files = glob.glob(os.path.join(path, "*_" + ds + ".csv"))
    total = []
    for filename in all_files:
        df = pd.read_csv(filename, header = 0, index_col = None)
        
        work_name = filename.split('_')[0]
        try:
            
            ind = work.index(work_name)
            total_lines += df.shape[0]
        except: continue
        
        df['work']=ind+1
        total.append(df)
    blah = pd.concat(total)
    assert total_lines == blah.shape[0]
    blah.to_csv(ds + '3.csv', index = False)
```
